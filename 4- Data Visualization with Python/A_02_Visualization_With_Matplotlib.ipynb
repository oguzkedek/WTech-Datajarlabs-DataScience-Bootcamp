In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import datetime
warnings.filterwarnings('ignore')

df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df.head()
```

```python
print((datetime.datetime.now() - datetime.timedelta(100)).strftime("%Y-%m-%d"))
```

```python
countries = ['Spain', 'France', 'Germany', 'Italy']
filtered_df = df.loc[(df['location'].isin(countries)) & (df['date'] > pd.to_datetime("2021-12-08"))]
```

```python
spain = filtered_df[df['location'] == 'Spain'].iloc[:100, :]
france = filtered_df[df['location'] == 'France'].iloc[:100, :]
germany = filtered_df[df['location'] == 'Germany'].iloc[:100, :]
italy = filtered_df[df['location'] == 'Italy'].iloc[:100, :]
```

```python
title_style = {'family': 'Arial', 'color': 'darkred', 'size': 20}
axis_style  = {'family': 'Arial', 'color': 'darkblue', 'size': 15}

plt.figure(figsize=(14,8))
plt.subplot(1, 2, 1)
plt.title("Total Cases", fontdict=title_style)
plt.xlabel('Dates', fontdict=axis_style)
plt.ylabel('Total Cases', fontdict=axis_style)
plt.plot(spain['date'], spain['total_cases'], c="red", marker="o", ms=2, ls="--", lw=5, label='Spain')
plt.plot(france['date'], france['total_cases'], c="blue", marker="*", ms=2, ls="-", lw=5, label='France')
plt.plot(germany['date'], germany['total_cases'], c="green", marker="+", ms=2, ls=":", lw=5, label='Germany')
plt.plot(italy['date'], italy['total_cases'], c="black", marker="s", ms=2, ls="-.", lw=5, label='Italy')
plt.legend(title='Countries', loc=2)
plt.xticks(rotation=30, fontsize=9, color='brown')

plt.subplot(1, 2, 2)
plt.title("Total Deaths", fontdict=title_style)
plt.xlabel('Dates', fontdict=axis_style)
plt.ylabel('Total Deaths', fontdict=axis_style)
plt.plot(spain['date'], spain['total_deaths'], c="red", marker="o", ms=2, ls="--", lw=5, label='Spain')
plt.plot(france['date'], france['total_deaths'], c="blue", marker="*", ms=2, ls="-", lw=5, label='France')
plt.plot(germany['date'], germany['total_deaths'], c="green", marker="+", ms=2, ls=":", lw=5, label='Germany')
plt.plot(italy['date'], italy['total_deaths'], c="black", marker="s", ms=2, ls="-.", lw=5, label='Italy')
plt.legend(title='Countries', loc=2)
plt.xticks(rotation=30, fontsize=9, color='brown')

plt.show()


```

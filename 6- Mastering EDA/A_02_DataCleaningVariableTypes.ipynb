```python
import pandas as pd
import numpy as np

movies = pd.read_csv('https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/movie.csv')
```

```python
def categorizer(budget):
    if budget >= 45000000:
        return 'high'
    elif (budget > 6000000) and (budget < 45000000):
        return 'medium'
    else:
        return 'low'
```

```python
movies['budget_group'] = movies['budget'].apply(categorizer)
display(movies.head())
movies.groupby("budget_group")["budget"].count()
```

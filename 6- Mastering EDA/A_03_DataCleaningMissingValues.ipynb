**(3)** Now it's time to consider the factor of time! Review your answer in question 2 and fill in the missing data based on the values observed during that year. For example, if you want to fill a value with an average value, calculate the average of that year.

```python
for column in states.columns[3:]:
    for year in states['YEAR'].unique():
        if (states[column].skew() <= -1) | (states[column].skew() >= 1):
            states[column].fillna(states[states['YEAR'] == year][column].median(), inplace=True)
        else:
            states[column].fillna(states[states['YEAR'] == year][column].mean(), inplace=True)
```
